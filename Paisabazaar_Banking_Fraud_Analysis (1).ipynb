## 🔹 **[3] DATA WRANGLING**

## Remove Duplicates


```python
df = df.drop_duplicates()
```

### Handle Missing Values


```python
df.isnull().sum()

df.fillna(method='ffill', inplace=True)
```

## Remove Outliers (IQR Method)

```python
Q1 = df['Annual_Income'].quantile(0.25)
Q3 = df['Annual_Income'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['Annual_Income'] >= lower) & (df['Annual_Income'] <= upper)]
```

```python
# Outlier Treatment for Multiple Columns using IQR Method

cols_to_treat = ['Num_of_Delayed_Payment', 'Outstanding_Debt', 'Interest_Rate',
                 'Num_Credit_Inquiries', 'Total_EMI_per_month']

print("Shape before outlier removal:", df.shape)

for col in cols_to_treat:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    before = df.shape[0]
    df = df[(df[col] >= lower) & (df[col] <= upper)]
    after = df.shape[0]
    print(f"{col}: removed {before - after} outlier rows")

print("Shape after outlier removal:", df.shape)
```

## Create New Column: Income Per Account

```python
df['Income_per_account'] = df['Annual_Income'] / df['Num_Bank_Accounts']
```

## Create Credit Score Category

```python
# ✅ FIXED: Credit_Score is already a string category ('Good', 'Standard', 'Poor')
# We just map it to a proper ordered column

credit_order = {'Poor': 1, 'Standard': 2, 'Good': 3}
df['Credit_Level'] = df['Credit_Score']  # direct copy (already categorical)
df['Credit_Score_Rank'] = df['Credit_Score'].map(credit_order)  # numeric rank for sorting/analysis

print("Credit Score Distribution:")
print(df['Credit_Score'].value_counts())
print()
print("Credit Level Unique Values:", df['Credit_Level'].unique())
```

```python
# Rule-Based Risk Flag
# A customer is HIGH RISK if they have:
# - More than 5 delayed payments AND
# - Credit utilization ratio above 70%

df['Risk_Flag'] = ((df['Num_of_Delayed_Payment'] > 5) &
                   (df['Credit_Utilization_Ratio'] > 70)).astype(int)

print("Risk Flag Distribution:")
print(df['Risk_Flag'].value_counts())
print()
total = len(df)
high_risk_pct = (df['Risk_Flag'] == 1).sum() / total * 100
print(f"High Risk Customers: {high_risk_pct:.2f}%")
print()

# How do risk flagged customers distribute across Credit Score?
print("Risk Flag vs Credit Score:")
print(pd.crosstab(df['Risk_Flag'], df['Credit_Score'], normalize='index').round(2) * 100)
```

## **6. customer segmentation [High risk vs low risk profile table]**

```python
# ============================================================
# [6] CUSTOMER SEGMENTATION ANALYSIS
# ============================================================

# Compare Good vs Poor credit customer profiles

seg_cols = ['Annual_Income', 'Monthly_Inhand_Salary', 'Outstanding_Debt',
            'Credit_Utilization_Ratio', 'Num_of_Delayed_Payment',
            'Num_of_Loan', 'Interest_Rate', 'Total_EMI_per_month',
            'Num_Credit_Inquiries', 'Monthly_Balance']

segment_table = df[df['Credit_Score'].isin(['Good', 'Poor'])]\
                  .groupby('Credit_Score')[seg_cols].mean().T.round(2)

segment_table.columns.name = None
segment_table['Difference (Good - Poor)'] = (segment_table['Good'] - segment_table['Poor']).round(2)

print("=" * 65)
print("     CUSTOMER SEGMENT COMPARISON: GOOD vs POOR CREDIT")
print("=" * 65)
print(segment_table.to_string())
print("=" * 65)
```

```python
# Visual: Side-by-side comparison of Good vs Poor customers

compare_metrics = ['Annual_Income', 'Outstanding_Debt',
                   'Num_of_Delayed_Payment', 'Credit_Utilization_Ratio',
                   'Interest_Rate', 'Monthly_Balance']

good_vals = df[df['Credit_Score'] == 'Good'][compare_metrics].mean()
poor_vals = df[df['Credit_Score'] == 'Poor'][compare_metrics].mean()

# Normalize for radar-like bar comparison
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
axes = axes.flatten()

for i, col in enumerate(compare_metrics):
    vals = [good_vals[col], poor_vals[col]]
    colors = ['#2ecc71', '#e74c3c']
    bars = axes[i].bar(['Good', 'Poor'], vals, color=colors, edgecolor='black', width=0.5)
    axes[i].set_title(col.replace('_', ' '), fontsize=11, fontweight='bold')
    axes[i].set_ylabel('Average Value')
    for bar, val in zip(bars, vals):
        axes[i].text(bar.get_x() + bar.get_width()/2,
                     bar.get_height() * 0.95,
                     f'{val:,.1f}', ha='center', va='top',
                     color='white', fontweight='bold', fontsize=9)

plt.suptitle('Good vs Poor Credit Customers — Key Metric Comparison',
             fontsize=15, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()

print("\n📌 Insight: Good credit customers earn more, carry less debt, delay payments less,")
print("and maintain higher monthly balances compared to Poor credit customers.")
```

## Risk score summary by occupation

```python
# Risk Summary by Occupation

occ_risk = df.groupby('Occupation').agg(
    Avg_Delayed_Payments = ('Num_of_Delayed_Payment', 'mean'),
    Avg_Credit_Utilization = ('Credit_Utilization_Ratio', 'mean'),
    Avg_Outstanding_Debt = ('Outstanding_Debt', 'mean'),
    Pct_Poor_Credit = ('Credit_Score', lambda x: (x == 'Poor').mean() * 100),
    Total_Customers = ('Customer_ID', 'count')
).round(2).sort_values('Pct_Poor_Credit', ascending=False)

print("=" * 75)
print("        RISK SUMMARY BY OCCUPATION (sorted by % Poor Credit)")
print("=" * 75)
print(occ_risk.to_string())
print("=" * 75)

# Visualize: % Poor Credit by Occupation
plt.figure(figsize=(14, 6))
bars = plt.barh(occ_risk.index, occ_risk['Pct_Poor_Credit'],
                color='#e74c3c', edgecolor='black')
plt.xlabel('% Customers with Poor Credit Score', fontsize=12)
plt.title('% Poor Credit Score by Occupation', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()
for bar, val in zip(bars, occ_risk['Pct_Poor_Credit']):
    plt.text(val + 0.2, bar.get_y() + bar.get_height()/2,
             f'{val:.1f}%', va='center', fontsize=9)
plt.tight_layout()
plt.show()

print("\n📌 Insight: Certain occupations show disproportionately high rates of Poor credit,")
print("which can help lenders apply occupation-aware risk assessment.")
```

# **🔹 [7] BUSINESS RECOMMENDATIONS**

Based on the analysis, the following actionable recommendations are proposed for Paisabazaar:

---

### 🚨 1. Implement a Rule-Based Risk Flagging System
Customers with **more than 5 delayed payments** AND **credit utilization above 70%** should be automatically flagged as High Risk before any loan approval. This catches the majority of Poor credit customers proactively.

---

### 💳 2. Use Payment Behaviour as a Key Screening Variable
Customers with **'High_spent_Small_value_payments'** behaviour show significantly higher Poor credit rates. Paisabazaar should weight this pattern more heavily during credit evaluation.

---

### 💰 3. Do Not Rely on Income Alone for Credit Approval
High-income customers can still have Poor credit scores due to high debt and poor payment habits. Income must always be evaluated alongside **Debt-to-Income Ratio** and **Credit Utilization**.

---

### 🏢 4. Apply Occupation-Aware Risk Thresholds
Certain occupations show higher rates of Poor credit. Paisabazaar can apply **tighter loan limits or higher interest rates** for high-risk occupation segments as a buffer.

---

### 📊 5. Target 'Standard' Customers for Credit Improvement Products
The 'Standard' credit group is the largest segment. Offering them **credit counseling, lower utilization alerts, and structured EMI plans** can help move them to 'Good' — reducing long-term risk and increasing customer lifetime value.

---

### ⏳ 6. Monitor Credit Inquiries as a Fraud Signal
Customers with **more than 5 credit inquiries in a short period** may be in financial distress or attempting credit fraud. These accounts should be flagged for manual review.
