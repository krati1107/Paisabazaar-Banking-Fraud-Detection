## 🔹 **[3] DATA WRANGLING**

## Remove Duplicates


```python
df = df.drop_duplicates()
```

### Handle Missing Values


```python
df.isnull().sum()

df.fillna(method='ffill', inplace=True)
```

## Remove Outliers (IQR Method)

```python
Q1 = df['Annual_Income'].quantile(0.25)
Q3 = df['Annual_Income'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['Annual_Income'] >= lower) & (df['Annual_Income'] <= upper)]
```

## Create New Column: Income Per Account

```python
df['Income_per_account'] = df['Annual_Income'] / df['Num_Bank_Accounts']
```

## Create Credit Score Category

```python
def credit_score_level(score):
    # This function is designed to categorize numerical scores.
    # However, the 'Credit_Score' column is already categorical (strings).
    # The error 'TypeError: '>=' not supported between instances of 'str' and 'int'' occurs
    # because it tries to compare string values (e.g., 'Good') with integers (750, 600).
    # Since 'Credit_Score' already holds the desired levels, we can directly use it.
    if score >= 750:
        return "Good"
    elif score >= 600:
        return "Standard"
    else:
        return "Poor"

df['Credit_Level'] = df['Credit_Score'] # Assign directly as Credit_Score is already categorical
```

## (Summary Points)

*  Remove outliers using IQR method (especially in Annual Income)
*   Define lower and upper limits to filter extreme values

*   Clean dataset by keeping only valid income range

*   Create new feature: Income per Account (Annual Income ÷ Bank Accounts)
*   Create new column for better analysis (feature engineering)


*   Convert Credit Score into categories:

         Good (≥ 750)
        Standard (600–749)
         Poor (< 600)



*  Use .apply() function for creating custom logic columns
Improve dataset quality for better analysis and modeling
